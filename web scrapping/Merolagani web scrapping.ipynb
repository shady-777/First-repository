```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup("https://merolagani.com/LatestMarket.aspx",'html.parser')
raw_data = requests.get(url)
raw_data
```

```python
soup = BeautifulSoup(raw_data.text,'lxml')
soup
```

```python
tablerows = soup.tbody.find_all('tr')
tablerows
```

```python
for tablerow in tablerows:
    print(tablerow.text)
```

```python

table_data = tablerow.find('tr',class_="increase-row odd")
print(table_data)
```

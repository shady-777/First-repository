```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup("https://www.worldometers.info/coronavirus/", 'html.parser')
raw_code = requests.get(url)
raw_code
```

```python
a = raw_code.text
```

```python
raw_code = BeautifulSoup(a, "lxml")
raw_code
```

```python
table_code = raw_code.table
table_code
```

```python
tags = table_code.find_all('tr')#find_all is a function to call all the data
tags
```

```python
for tag in tags:
    print(tag.text)
```

```python
raw_data=[]
for tag in tags:
    x = tag.text.split("\n")
    if x[1] !='':
        raw_data.append(x[1:])
    
    
raw_data
```

```python
import csv
with open('covid_data.csv','w') as file:
    x = csv.writer(file)
    x.writerows(raw_data)
    file.close()
```

```python
import pandas as pd
df = pd.read_csv('covid_data.csv',encoding ='ISO-8859-1')
df
```

```python
import plotly.graph_objects as go
country=df['Country,Other'][0:10]
total_cases = df['TotalCases'][0:10]
new_total_cases = []
for i in total_cases:
    new_total_cases.append(int(i.replace(',','')))

fig = go.Figure([go.Bar(x=country, y=new_total_cases)])
fig.show()
```

```python
import plotly.graph_objects as go



fig = go.Figure(data=[go.Pie(labels=country, values=new_total_cases)])
fig.show()
```

```python
import plotly.graph_objects as go
new_recovered =df['TotalRecovered'][0:10]
new_total_recovered = []
for i in new_recovered:
    new_total_recovered.append(int(i.replace(',','')))

fig = go.Figure(data=[
    go.Bar(name='Total cases', x=country, y=new_total_cases),
    go.Bar(name='Total Recovered', x=country, y=new_total_recovered)
#     go.Bar(name='Total Deaths', x=country, y=new_total_recovered)
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```
